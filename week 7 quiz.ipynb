```python
import pandas as pd
import numpy as np

df = pd.read_csv("titanic.csv")
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder

class_labels = LabelEncoder()

df['Sex'] = class_labels.fit_transform(df['Sex'].values)
df
```

```python
newtitanicdf = df[['Survived','Age','SibSp','Parch','Fare','Sex']].copy()
df_corr = df[['Survived']]
newtitanicdf.corr()
```

```python
from statsmodels.formula.api import logit
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression


X = df[['Sex','Pclass']]
# x is what we are using to predict y
#X = np.array(df['Gender']).reshape(-1,1)

# y is what we are predicting
y = np.array(df['Survived']).reshape(-1,1)

#1. train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
regression = LogisticRegression(random_state=42).fit(X_train, y_train)
y_predicted = regression.predict(X_test)
```

```python
y_pred = regression.predict(X_test)
y_pred
```
