```python
# 5
import time
time1 = time.perf_counter()
age_bins = [0,12,19,64,100]
age_labels = ['Children', 'Teens','Adults','Seniors']
age_bins = pd.cut(df['Age'], age_bins, labels=age_labels)
df.insert(6, "Age_bins", age_bins)
time2 = time.perf_counter()
print(time2 - time1)
df.head()
```

```python
#5
time_lambda1 = time.perf_counter()
age_bins_df = df['Age'].apply(lambda x: 'Children' if x <= 12 else 'Teens' if x <= 19 else 'Adults' if x <= 64 else 'Seniors')
time_lambda2 = time.perf_counter()
print(time_lambda2 - time_lambda1)
```

```python
Children_df = df.loc[df["Age_bins"] == "Children"]
Teens_df = df.loc[df["Age_bins"] == "Teens"]
Adults_df = df.loc[df["Age_bins"] == "Adults"]
Seniors_df = df.loc[df["Age_bins"] == "Seniors"]
```
